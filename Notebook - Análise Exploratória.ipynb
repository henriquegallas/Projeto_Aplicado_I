```python
import pandas as pd
```

```python
data_area_colhida = pd.read_csv('area-colhida-hectares-2021.csv', sep=';')
data_area_plantada = pd.read_csv('area-plantada-ou-destinada-a-colheita-hectares-2021.csv', sep=';')
data_efetivo_rebanhos = pd.read_csv('efetivo-rebanhos-por-cabecas-2021.csv', sep=';')
data_producao_origem_animal = pd.read_csv('producao-origem-animal-por-tipo-2021.csv', sep=';')
data_quantidade_produzida = pd.read_csv('quantidade-produzida-tonelada-2021.csv', sep=';')
data_ranking_agricultura = pd.read_csv('ranking-agricultura-valor-da-producao-brasil.csv')
data_ranking_pecuaria = pd.read_csv('ranking-pecuaria-rebanhos-brasil.csv')
data_rendimento_medio_producao = pd.read_csv('rendimento-medio-producao-kg-hectare-2021.csv', sep=';')
data_valor_producao = pd.read_csv('valor-producao-mil-reais-2021.csv', sep=';')
```
